# Module: Data Visualization with Seaborn

Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
To get started with Seaborn, we first need to import it:


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing required libraries
import numpy as np
import pandas as pd

# importing matplotlib
import matplotlib.pyplot as plt

# display plots in the notebook itself
%matplotlib inline
```

```python
# read the dataset
url='https://raw.githubusercontent.com/ubaid-shah/Python-Advance/main/Datasets/bigmart_data.csv'
data_BM = pd.read_csv(url)
# drop the null values
bigmart_data = data_BM.dropna(how="any")
# view the top results
bigmart_data.head()
```

## 2. Distribution Plots

Distribution plots are used for visualizing the distribution of a dataset.

```python
sns.histplot(bigmart_data['Item_Outlet_Sales'], bins=30, kde=True )
```

## 3. Categorical Plots

Categorical plots are used for visualizing categorical data.

```python
sns.barplot(x='Outlet_Type', y='Item_Outlet_Sales', data=bigmart_data)
```

## 4. Matrix Plots/Correlation

Matrix plots are used to display data in matrix form.

```python
correlation = bigmart_data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
```

## 5. Pair Plots

Pair plots are used for pairwise relationship visualization.

```python
subset_data = bigmart_data[['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Item_Outlet_Sales']]
sns.pairplot(subset_data)
```

## 6. Regression Plots

Regression plots are used to visualize the relationship between two variables.

```python
sns.regplot(x='Item_MRP', y='Item_Outlet_Sales', data=bigmart_data, scatter_kws={'s': 10})
```

## 7. Styling and Themes

Demonstrates how to customize the appearance of plots in Seaborn.

```python
sns.set_style('whitegrid')
sns.boxplot(x='Outlet_Type', y='Item_Outlet_Sales', data=bigmart_data)
```
